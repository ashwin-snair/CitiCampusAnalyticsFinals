## Importance of categorical metrics 

```python
def goodmanKruskalLambda(Y, X):
    '''
    X : Independent nominal variable (numpy array/pandas series)
    Y : Dependent nominal variable (numpy array/pandas series)
    
    (Lambda with Y (row variable) as the dependent variable is calculated)
    
    returns:
    ---------
    gklambda 
    '''
    contingency_table = pd.crosstab(Y, X).values
    row_total = np.sum(contingency_table, axis=0)
    
    S = np.sum(np.max(contingency_table, axis=1)) # sum of the highest cell count for each row
    R = np.max(row_total)                         # the highest row total
    N = np.sum(row_total)                         # the total of all the cell counts
    gKLambda = (S-R)/(N-R)
    
    return gKLambda


def phiCoefficient(X, Y):
    '''
    X : Binary variable (numpy array/pandas series)
    Y : Binary variable (numpy array/pandas series)
    
    returns:
    ---------
    phiCoeff
    '''
    contingency_table = pd.crosstab(X, Y).values

    a = contingency_table[0][0]
    b = contingency_table[0][1]
    c = contingency_table[1][0]
    d = contingency_table[1][1]

    phi = (a*d-b*c)
    den = sqrt(a+b)
    den *= sqrt(c+d)
    den *= sqrt(a+c)
    den *= sqrt(b+d)   # This is done in multiple steps to avoid overflow
    phi /= den   
    return phi


def chi2_and_stats(X, Y, print_chi2 = True):
    '''
    X : Nominal variable (numpy array/pandas series)
    Y : Nominal variable (numpy array/pandas series)
    
    returns:
    ---------
    (V, V_bc) : Crammer's V (with bias correction)
    (T, T_bc) : Tschuprow's T (with bias correction)
    C         : Contingency Coefficient 
    '''
    contingency_table = pd.crosstab(X, Y).values  # Observed values
    
    chi2, p, dof, expected_values = chi2_contingency(contingency_table)
    
    if print_chi2 == True:
        print('Chi^2 test results:')
        print('chi^2 stat = {0}, p = {1}'.format(chi2, p))
        
    n = np.sum(contingency_table)
    phi = chi2/n    
    k = contingency_table.shape[0]
    r = contingency_table.shape[1]
    
    V = sqrt(phi/(min(k,r)-1))
    T = sqrt(phi/sqrt((k-1)*(r-1)))
    
    # bias correction
    phiBar = max(0, phi**2 - (k-1)*(r-1)/(n-1))
    kBar = k - ((k-1)**2)/(n-1)
    rBar = r - ((r-1)**2)/(n-1)
    
    V_bc = sqrt(phiBar/(min(kBar,rBar)-1))
    T_bc = sqrt(phiBar/sqrt((kBar-1)*(rBar-1)))
    
    C = sqrt(chi2/(n+chi2))
    
    return (V, V_bc), (T, T_bc), C
        
```

### Do companies use different bank accounts for different types of transactions?

#check again

```python
t_types = train.groupby('rem_company_id_dummy')['txn_type'].nunique() # number of types of transactions done by a company
no_of_bankacc = train.groupby('rem_company_id_dummy')['rem_act_no_dummy'].nunique() # number unique bank accounts a company owns

# Correlation
print(np.corrcoef(t_types.values, no_of_bankacc.values))


# chi2 independence test
obs_bt = pd.crosstab(no_of_bankacc,t_types)
print(obs_bt)
chi2_contingency(obs_bt.values)

# Conclusion: No these columns are not related
```
