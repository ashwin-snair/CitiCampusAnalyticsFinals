```python
train['dd'] = train.apply(lambda x: x['transaction_val_dt'].day, axis=1)

rem_gps = train.groupby('rem_act_no_dummy')
payroll_avg = rem_gps['payroll_ind'].mean()
weights = np.ones_like(payroll_avg) / len(payroll_avg)
plt.hist(payroll_avg, bins=20, weights=weights)
plt.title('Homogenity Distribution')
plt.show()

homo_rems = sum((rem_gps['payroll_ind'].mean()==1) | (rem_gps['payroll_ind'].mean()==0))
print('Number of homogenous solution remanies:', homo_rems)
```

```python
rem_names = list(rem_gps.groups.keys())
```

```python
def fn(x):
    n_inds = len(np.unique(x))
    if n_inds == 1 or 'Unknown' not in x:
        return n_inds
    else:
        return n_inds-1
    
n_ind = rem_gps['rem_company_ind'].unique().apply(fn)
```

```python
multi_ind = rem_gps['rem_company_ind'].unique().apply(fn)
inds = list(multi_ind[multi_ind>1].index)
print('Transactions of rem accounts with more than 1 industry', len(train[train['rem_act_no_dummy'].isin(inds)]))
```

```python
groupDF['rem_act_no_dummy'] = pd.Series(groupDF.index).apply(lambda x:int(x.split(',')[0][1:])).values
```

```python
def get_n_ind(x):
    global multi_ind
    rem = x['rem_act_no_dummy']
    return multi_ind[rem]

groupDF['rem_n_ind'] = groupDF.apply(get_n_ind, axis=1)
```

```python
rem_dd_means = rem_gps['dd'].mean()
rem_dd_vars = rem_gps['dd'].var(ddof=0)

def get_dd_props(x):
    global rem_dd_means, rem_dd_vars
    rem = x['rem_act_no_dummy']
    return abs(rem_dd_means[rem]-x['dd_mean']), x['dd_var']/rem_dd_vars[rem]

dd_props = groupDF.apply(get_dd_props, axis=1)
groupDF['dd_diff_rem'], groupDF['dd_var_ratio_rem'] = dd_props.str
```

```python
rem_txn_means = rem_gps['txn_amt'].mean()
rem_txn_vars = rem_gps['txn_amt'].var(ddof=0)

def get_txn_props(x):
    global rem_txn_means, rem_txn_vars
    rem = x['rem_act_no_dummy']
    return rem_txn_means[rem]-x['txn_amt_mean'], x['txn_amt_var']/rem_txn_vars[rem]

txn_props = groupDF.apply(get_txn_props, axis=1)
groupDF['txn_amt_diff_rem'], groupDF['txn_amt_var_ratio_rem'] = txn_props.str
```

```python
def getAvgTxnPerDay(x):
    return len(x)/x['transaction_val_dt'].nunique()

avgTxnPerDay = rem_gps.apply(getAvgTxnPerDay)

def get_txn_per_day_ratio(x):
    global avgTxnPerDay
    rem = x['rem_act_no_dummy']
    return x['avgTxnPerDay']/avgTxnPerDay[rem]

txn_per_day_ratio = groupDF.apply(get_txn_per_day_ratio, axis=1)
groupDF['txnPerDay_ratio_rem'] = txn_per_day_ratio
```

```python
def getGapVar(x):
    if len(x)>1:
        return x['transaction_val_dt'].sort_values().diff()[1:].apply(lambda x: x.days).var(ddof=0)
    else:
        return 0
    
delta = (rem_gps['transaction_val_dt'].max()-rem_gps['transaction_val_dt'].min()).apply(lambda x: x.days)
avgTxnGap = delta/(rem_gps.count()['payroll_ind']-1)
varTxnGap = rem_gps.apply(getGapVar)
```

```python
def get_txn_gaps(x):
    global avgTxnGap, varTxnGap
    rem = x['rem_act_no_dummy']
    return avgTxnGap[rem]-x['avg_txn_time'], x['varTxnGap']/varTxnGap[rem]

txn_gap_info = groupDF.apply(get_txn_gaps, axis=1)
groupDF['txnGap_diff_rem'], groupDF['txnGap_var_ratio_rem'] = txn_gap_info.str
```

# Inserting the columns from my other Dataset

```python
def import_Train():
    Train = pd.read_csv('Train.csv')
    Train['Year'] = Train['transaction_val_dt'].apply(lambda x:int(str(x).split('-')[0]))
    Train['Month'] = Train['transaction_val_dt'].apply(lambda x:int(str(x).split('-')[1]))
    Train['Day'] = Train['transaction_val_dt'].apply(lambda x:int((str(x).split('-')[2].split(' ')[0])))
    return Train
```

```python
def create_train(Train):
    train = Train[~Train.index.duplicated(keep='first')]
    train['ID'] = train.index
    return train
```

```python
def reindex_train(Train):
    Train = Train.astype({"rem_act_no_dummy": str, "bene_act_no_dummy": str})
    Train['Index'] = Train['rem_act_no_dummy'].str.cat(Train['bene_act_no_dummy'],sep=", ")
    Train['Index'] = Train['Index'].apply(lambda x: '('+x+')')
    Train = Train.astype({"rem_act_no_dummy": int, "bene_act_no_dummy": int})
    Train = Train.drop(['yearmonth'], axis = 1)
    Train = Train.set_index('Index')
    return Train
```

```python
def create_txn_type_num(train, Train):                          
    '''
    train : groupDF
    Train : original train dataset with reindexing
    
    Changes txn_type -> txn_type for the pair (lies b/w 0 to 1)
    
    '''
    
    Unique_txn_type = Train.groupby('Index').txn_type.unique()
    temp = pd.DataFrame(Unique_txn_type)
    temp['length'] = temp['txn_type'].apply(lambda x:len(x))
    multiple_txn_type = temp[temp.length>1]
    train.replace({'DOMESTIC':1,'CROSS BORDER':0},inplace=True)
    ratios = []
    multiple_index = multiple_txn_type.index
    for i in multiple_index:
        num = 0
        den = 0
        # print(Train.loc[i,'txn_type'].value_counts())
        ratio = Train.loc[i,'txn_type'].value_counts()[0]/(Train.loc[i,'txn_type'].value_counts()[1]+Train.loc[i,'txn_type'].value_counts()[0])
        ratios.append(ratio)
        train.loc[i,'txn_type'] = ratio
    train = train.rename(columns={'txn_type': 'txn_type_num'})
    return train
```

```python
def create_rem_company_ind(train,Train):
    '''
    train : groupDF
    Train : original train dataset with reindexing
    
    pairs with 2 -> by removing unknowns. 
    
    '''    
    
    company_ind = pd.DataFrame(Train.groupby('Index').rem_company_ind.unique())
    company_ind.iloc[:,0] = company_ind.iloc[:,0].apply(lambda x:np.sort(x)[0])
    train['rem_company_ind'] = train.ID.apply(lambda x: company_ind.loc[x,'rem_company_ind'])
    return train
```
